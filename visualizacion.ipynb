```python

# Assuming df_primaria and df_secundaria are your DataFrames
ASIGNATURAS_PRIMARIA = ['PCAT', 'PCAST', 'PMAT', 'PANG', 'PMED', 'NATURALESA']
ASIGNATURAS_SECUNDARIA = ['PCAT', 'PCAST', 'PMAT', 'PANG', 'PCIEN', 'NATURALESA']

# Melt the DataFrames to reshape them
melted_df_primaria = pd.melt(df_primaria[:], id_vars=['NATURALESA'], value_vars=ASIGNATURAS_PRIMARIA, var_name='Subject', value_name='Value')
melted_df_secundaria = pd.melt(df_secundaria[:], id_vars=['NATURALESA'], value_vars=ASIGNATURAS_SECUNDARIA, var_name='Subject', value_name='Value')

melted_df_primaria['NATURALESA'] = melted_df_primaria['NATURALESA'].astype(str)
melted_df_secundaria['NATURALESA'] = melted_df_secundaria['NATURALESA'].astype(str)

# Especificar el orden deseado
order = ['Privada', 'Pública']

# Convertir 'NATURALESA' a tipo categoría con el orden deseado
melted_df_primaria['NATURALESA'] = pd.Categorical(melted_df_primaria['NATURALESA'], categories=order, ordered=True)
melted_df_secundaria['NATURALESA'] = pd.Categorical(melted_df_secundaria['NATURALESA'], categories=order, ordered=True)


# Ordenar por la columna 'NATURALESA'
melted_df_primaria[melted_df_primaria['Subject']=='PCAT'] = melted_df_primaria[melted_df_primaria['Subject']=='PCAT'].sort_values(by='NATURALESA')
melted_df_secundaria[melted_df_secundaria['Subject']=='PCAT'] = melted_df_secundaria[melted_df_secundaria['Subject']=='PCAT'].sort_values(by='NATURALESA')


# Crear un diccionario para mapear nombres de asignaturas
subject_mapping = {
    'PCAT': 'Catalan',
    'PCAST': 'Spanish',
    'PMAT': 'Mathematics',
    'PANG': 'English',
    'PMED': 'Science',
    'PCIEN': 'Science'  
}
# Mapear nombres de asignaturas en las columnas 'Subject'
melted_df_primaria['Subject'] = melted_df_primaria['Subject'].map(subject_mapping)
melted_df_secundaria['Subject'] = melted_df_secundaria['Subject'].map(subject_mapping)

# Create a dictionary to map original class names to desired labels
class_labels = {'Privada': 'Private', 'Pública': 'Public'}

# Map the original class names to desired labels in the 'Naturaleza' column
melted_df_primaria['NATURALESA'] = melted_df_primaria['NATURALESA'].map(class_labels)
melted_df_secundaria['NATURALESA'] = melted_df_secundaria['NATURALESA'].map(class_labels)
```

```python
# Create subplots
sns.set(style="whitegrid", palette='viridis', rc={'legend.framealpha': 0.7})
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Create boxplots for primaria
box= sns.boxenplot(data=melted_df_primaria, x="Subject", y='Value', hue='NATURALESA', palette='viridis', ax=axes[0],linewidth=.3, legend=False)
axes[0].set_xlabel('Subject')
axes[0].set_ylabel('Grades in Sixth Grade of Primary Education')


# Create boxplots for secundaria
sns.boxenplot(data=melted_df_secundaria, x="Subject", y='Value', hue='NATURALESA', palette='viridis', ax=axes[1],linewidth=.3)
axes[1].set_xlabel('Subject')
axes[1].set_ylabel('Grades in Fourth Grade of Secondary Education')


# Show the legend for both plots
# axes[0].legend(title='',loc='lower left')
axes[1].legend(bbox_to_anchor=(1, 0.55), loc='upper left')


# Adjust layout
# plt.tight_layout()

# Show the plots
plt.show()
```

```python


# Assuming df_primaria and df_secundaria are your DataFrames
ASIGNATURAS_PRIMARIA = ['PCAT', 'PCAST', 'PMAT', 'PANG', 'PMED', 'HÀBITAT']
ASIGNATURAS_SECUNDARIA = ['PCAT', 'PCAST', 'PMAT', 'PANG', 'PCIEN', 'HÀBITAT']

# Melt the DataFrames to reshape them
melted_df_primaria = pd.melt(df_primaria[:], id_vars=['HÀBITAT', 'NATURALESA'], value_vars=ASIGNATURAS_PRIMARIA, var_name='Subject', value_name='Value')
melted_df_secundaria = pd.melt(df_secundaria[:], id_vars=['HÀBITAT', 'NATURALESA'], value_vars=ASIGNATURAS_SECUNDARIA, var_name='Subject', value_name='Value')

melted_df_primaria['HÀBITAT'] = melted_df_primaria['HÀBITAT'].astype(str)
melted_df_secundaria['HÀBITAT'] = melted_df_secundaria['HÀBITAT'].astype(str)

population_mapping = {
    'Fins a 10000': 'Up to 10,000',
    'De 10001 a 100000': 'From 10,001 to 100,000',
    'Més de 100000': 'More than 100,000', 
}
melted_df_primaria['HÀBITAT']=melted_df_primaria['HÀBITAT'].map(population_mapping)
melted_df_secundaria['HÀBITAT']=melted_df_secundaria['HÀBITAT'].map(population_mapping)

# Especificar el orden deseado
order = ['Up to 10,000','From 10,001 to 100,000',  'More than 100,000']

# Convertir 'HÀBITAT' a tipo categoría con el orden deseado
melted_df_primaria['HÀBITAT'] = pd.Categorical(melted_df_primaria['HÀBITAT'], categories=order, ordered=True)
melted_df_secundaria['HÀBITAT'] = pd.Categorical(melted_df_secundaria['HÀBITAT'], categories=order, ordered=True)

# Ordenar por la columna 'HÀBITAT'
melted_df_primaria[melted_df_primaria['Subject']=='PCAT'] = melted_df_primaria[melted_df_primaria['Subject']=='PCAT'].sort_values(by='HÀBITAT')
melted_df_secundaria[melted_df_secundaria['Subject']=='PCAT'] = melted_df_secundaria[melted_df_secundaria['Subject']=='PCAT'].sort_values(by='HÀBITAT')

# Crear un diccionario para mapear nombres de asignaturas
subject_mapping = {
    'PCAT': 'Catalan',
    'PCAST': 'Spanish',
    'PMAT': 'Mathematics',
    'PANG': 'English',
    'PMED': 'Science',
    'PCIEN': 'Science'  
}
# Mapear nombres de asignaturas en las columnas 'Subject'
melted_df_primaria['Subject'] = melted_df_primaria['Subject'].map(subject_mapping)
melted_df_secundaria['Subject'] = melted_df_secundaria['Subject'].map(subject_mapping)


# Show the first few rows of the melted DataFrames
print("Melted DataFrame - Primaria:")
print(melted_df_primaria.head())

print("\nMelted DataFrame - Secundaria:")
print(melted_df_secundaria.head())
```

```python
from scipy.stats import ttest_ind

ASIGNATURAS = ['PCAT', 'PCAST', 'PMAT', 'PANG', 'PFRAN', 'PMED']

# Crear subconjuntos para privada y pública
privada_data = df_primaria[df_primaria['NATURALESA'] == 'Privada']
publica_data = df_primaria[df_primaria['NATURALESA'] == 'Pública']

# Realizar la prueba t de Student para cada asignatura
for asignatura in ASIGNATURAS:
    # Eliminar valores perdidos
    privada_data_asignatura = privada_data[asignatura].dropna()
    publica_data_asignatura = publica_data[asignatura].dropna()

    # Calcular medias y varianzas
    media_privada = privada_data_asignatura.mean()
    media_publica = publica_data_asignatura.mean()
    varianza_privada = privada_data_asignatura.var()
    varianza_publica = publica_data_asignatura.var()

    # Realizar la prueba t de Student
    _, p_value = ttest_ind(privada_data_asignatura, publica_data_asignatura, equal_var=False)

    # Imprimir resultados
    print(f'Asignatura: {asignatura}')
    print(f'Media privada: {media_privada:.2f}, Varianza privada: {varianza_privada:.2f}')
    print(f'Media pública: {media_publica:.2f}, Varianza pública: {varianza_publica:.2f}')
    print(f'P-value para {asignatura}: {p_value}')

    # Puedes agregar una condición para interpretar el resultado
    if p_value < 0.05:
        print(f'Diferencia significativa para {asignatura} entre privada y pública\n')
    else:
        print(f'No hay diferencia significativa para {asignatura} entre privada y pública\n')
```

```python
from scipy.stats import mannwhitneyu

ASIGNATURAS = ['PCAT', 'PCAST', 'PMAT', 'PANG', 'PFRAN', 'PMED']

# Crear subconjuntos para privada y pública
privada_data = df_primaria[df_primaria['NATURALESA'] == 'Privada']
publica_data = df_primaria[df_primaria['NATURALESA'] == 'Pública']

# Realizar la prueba de Mann-Whitney U para cada asignatura
for asignatura in ASIGNATURAS:
    # Eliminar valores perdidos
    privada_data_asignatura = privada_data[asignatura].dropna()
    publica_data_asignatura = publica_data[asignatura].dropna()

    # Realizar la prueba de Mann-Whitney U
    _, p_value = mannwhitneyu(privada_data_asignatura, publica_data_asignatura, alternative='two-sided')

    # Imprimir resultados
    print(f'Asignatura: {asignatura}')
    print(f'P-value para {asignatura}: {p_value}')

    # Puedes agregar una condición para interpretar el resultado
    if p_value < 0.05:
        print(f'Diferencia significativa para {asignatura} entre privada y pública\n')
    else:
        print(f'No hay diferencia significativa para {asignatura} entre privada y pública\n')
```

## Temporal evolution

```python
def plotyears(df, curso, YEAR, ASIGNATURAS, ax, parametro, subplot=False, years=None, legend=True):
    dfs = []

    for year in df[YEAR].unique():
        for asignatura in ASIGNATURAS:
            for naturaleza in df['NATURALESA'].unique():
                subset = df[asignatura][(df[YEAR] == year) & (df['NATURALESA'] == naturaleza)].dropna()
                median = subset.median()
                var=subset.var()

                # Create a DataFrame with the information for the current plot
                df_temp = pd.DataFrame({
                    'ASIGNATURA': [asignatura],
                    'YEAR': [year],
                    'Median': [median],
                    'Variance':[var],
                    'Naturaleza': [naturaleza],
                    'Grade': [curso]
                })

                # Add the current DataFrame to the list
                dfs.append(df_temp)

    # Concatenate all the DataFrames into one
    df_resultado = pd.concat(dfs, ignore_index=True)
    df_resultado['Naturaleza'] = df_resultado['Naturaleza'].astype(str)

    # Specify the desired order
    order = ['Privada', 'Pública']

    # Convert 'NATURALESA' to a categorical type with the desired order
    df_resultado['Naturaleza'] = pd.Categorical(df_resultado['Naturaleza'], categories=order, ordered=True)

    # Create a dictionary to map original class names to desired labels
    class_labels = {'Privada': 'Private', 'Pública': 'Public'}

    # Map the original class names to desired labels in the 'Naturaleza' column
    df_resultado['Naturaleza'] = df_resultado['Naturaleza'].map(class_labels)

    sns.set(style="whitegrid", palette='viridis', rc={'legend.framealpha': 0.5})
    
    # Main plot
    
    sns.lineplot(x="YEAR", y= parametro, hue="Naturaleza", data=df_resultado,ax=ax, palette='viridis',legend=legend)
    ax.set_xlabel('Year')
    ax.set_ylabel(f'{parametro} in {curso}')

    
    # Inset plot
    if subplot:
        inset_axes = ax.inset_axes([0.1, 0, 0.5, 0.6])
        sns.lineplot(x="YEAR", y= parametro, hue="Naturaleza", data=df_resultado[df_resultado['YEAR'].isin(years)], ax=inset_axes,legend=legend, palette=palette)
        inset_axes.set_xlabel(None)
        inset_axes.set_ylabel(None)
        inset_axes.legend_.remove()

```
