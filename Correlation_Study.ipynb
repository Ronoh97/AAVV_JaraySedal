```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import import_ipynb
import Plot_Functions as pf
```

```python
primary_df = pd.read_csv('Data\Avaluaci__de_sis__d_educaci__prim_ria.csv')
secondary_df = pd.read_csv('Data\Avaluaci__de_quart_d_Educaci__Secund_ria_Obligat_ria_20231112.csv')

primary_df = primary_df.drop_duplicates()
secondary_df = secondary_df.drop_duplicates()
replace_fields = ['PFRAN','PFRAN_CO','PFRAN_CL','PFRAN_EE']
secondary_df[replace_fields]= secondary_df[replace_fields].replace(1,np.nan)
```

```python
fields_1 = list(['PCAT','PCAST','PANG','PFRAN','PMAT','PMED'])
fields_2 = list(['PCAT','PCAST','PANG','PFRAN','PMAT','PCIEN'])

title = 'Títol'
pf.plot_violin_double(primary_df, secondary_df, fields_1, fields_2, title)
```

Here we can see the distribution global marks of the differnet subjects for both primary and secondary school.
We can see that some spaced peaks for spanish marks in primary and science marks in secondary, probably due to a low granularity in the tests from which they are averaged.
There is also a noticeable peak around 50 for english that may be due to a tendency of examinators to have students pass when there is some doubt.

```python
fields_1 = list(['PCAST','PCAST_EE','PCAST_CL','PMED','PMED_COMP1','PMED_COMP2','PMED_COMP3'])
fields_2 = list(['PCAST','PCAST_EE','PCAST_CL','PCIEN','PCIEN_COMP1','PCIEN_COMP2','PCIEN_COMP3'])

title = 'Títol'
pf.plot_violin_double(primary_df, secondary_df, fields_1, fields_2, title)
```

```python
fields_1 = list(['PCAT','PCAST','PANG','PFRAN','PMAT','PMED'])
fields_2 = list(['PCAT','PCAST','PANG','PFRAN','PMAT','PCIEN'])

title = 'Títol'
pf.plot_corr_double(primary_df, secondary_df, fields_1, fields_2, title)
```

This graph show the linear correlation between global marks of the different subjects.
We can see that all correlations are positive as expected. However, for secondary french has an extremely low correlation value with english.
We can also see that the highest correlation is that between catalan and spanish in primary.
It is also noticeable that marks in catalan are in primary school the better indicator of overall performance, this can have multiple interpretations, from the socio-economical distribution of people according to their primary language to the effects the fact that most subjects are taught in catalan, further data should be considered in order to reach a definite conclusion. 
THis effect disappears in secondary school.
